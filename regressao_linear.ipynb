```python
import pandas as pd
```

```python
dados = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv')
```

```python
covid_brasil = dados.iloc[30, :]
```

```python
covid_brasil
```

```python
data = pd.DataFrame(covid_brasil)
data
```

```python
data = data[-50:]
```

```python
data
```

```python
data.index == '5/25/21'
```

```python
for i in range(len(data)):
  lista = data.index == '5/25/21'

for i, v in enumerate(lista):
  if v == True:
    print(data.iloc[i, 0])
```

```python
data.reset_index(drop=True, inplace=True)
```

```python
data
```

```python
X = data.index.values.reshape((-1, 1))
X
```

```python
y = data.iloc[:, 0]
```

```python
y
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(X, y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
modelo = LinearRegression()
```

```python
modelo.fit(X, y)
```

```python
linha = modelo.predict(X)
```

```python
linha
```

```python
plt.plot(linha, color='red')
plt.scatter(X, y)
plt.figure()
```

```python
print(f'R² = {modelo.score(X, y)}')
```
